```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("shopping_trends.csv", index_col="Customer ID")
data
```

```python
data.info()
```

```python
data.describe()
```

```python
data.describe(include="object")
```

```python
slovnik_pohlavi = {"Female": 0, "Male": 1}
# Convert 'Gender' column to string type to ensure consistency
data["Gender"] = data["Gender"].astype(str)
# Applying the mapping to the 'Gender' column
data["Gender"] = data["Gender"].apply(lambda x: slovnik_pohlavi.get(x, x))
data["Gender"]
```

```python
slovnik_sub = {"No": 0, "Yes": 1}
data["Subscription Status"] = data["Subscription Status"].apply(lambda x: slovnik_sub[x])
data["Subscription Status"]
```

```python
slovnik_discount = {"No": 0, "Yes": 1}
data["Discount Applied"] = data["Discount Applied"].apply(lambda x: slovnik_discount[x])
data["Discount Applied"]
```

```python
slovnik_promo_code = {"No": 0, "Yes": 1}
data["Promo Code Used"] = data["Promo Code Used"].apply(lambda x: slovnik_promo_code[x])
data["Promo Code Used"]
```
